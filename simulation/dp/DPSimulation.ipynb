## Simulation

```python
# sim_name: str = "a1p19050_b1_D0p04_η1_x250_y250_Δx1_Δt0p1"
# info_path: list[str] = [pardir, pardir, "experiments", sim_name]
# sim_name: str = "a1p19050_b1_D0p04_η1_x31_y31_Δx1_Δt0p1"
# sim_name: str = "a0p28000_b1_D1p00_η1_x100_y50_Δx1_Δt0p1"
sim_name: str = "a0p26850_b1_D1p00_η1_x100_y50_Δx1_Δt0p1"
info_path: list[str] = [pardir, pardir, "demos", sim_name]
info: dict
_, info = read_info(info_path, dplvn)
_
```

```python
simulation = Simulation(
    name=sim_name, path=info_path, info=info, 
    do_snapshot_grid=True, do_verbose=True,
)
simulation.misc["name"]
```

```python
simulation.initialize()
```

```python
simulation.analysis["n_epochs"]
```

```python
simulation.run()
```
